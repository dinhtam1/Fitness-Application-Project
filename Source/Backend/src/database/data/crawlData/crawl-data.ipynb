```python
import pandas as pd
import numpy as np
import requests
import json
from bs4 import BeautifulSoup
import threading
import math
```

```python
url_web = "https://musclewiki.com/"
url_muscles = "https://musclewiki.com/newapi/muscle/muscles/"
url_equipment = "https://musclewiki.com/newapi/exercise/categories/?enable=true"

muscles = requests.get(url_muscles)

muscles = json.loads(muscles.text)

muscles = muscles['results']
len_muscles= len(muscles)

url_muscles_id = []
url_muscles_name = []

for i in range(len_muscles):
    url_muscles_id.append(muscles[i]['id'])
    url_muscles_name.append(muscles[i]['url_name'])

muscles = pd.DataFrame({'id': url_muscles_id, 'name': url_muscles_name})

equipment = requests.get(url_equipment)
equipment = json.loads(equipment.text)
equipment = equipment['results']

len_equipment= len(equipment)
url_equipment_id = []
url_equipment_name = []

for i in range(len_equipment):
    url_equipment_id.append(equipment[i]['id'])
    url_equipment_name.append(equipment[i]['name'])

equipmen = pd.DataFrame({'id': url_equipment_id, 'name': url_equipment_name})

gender = ['male', 'female']

ordering = ['-weight', '+weight']

# category = 24
status = "Published"
# muscles = 2

def requests_api(url):
    response = requests.get(url)
    if (response.status_code != 200):
        return False
    response = json.loads(response.text)
    return response

```

```python
df = pd.DataFrame(columns=['musclesId', 'musclesName', 'equipmentId', 'equipmentName','image' ,'gender', "ordering", "video_center", "video_side"])
lenDF = 0
lenKT = 0

def get_video(len_muscles, len_equipment):
    global lenDF
    global lenKT
    print("len_muscles: ", len_muscles)
    print("len_equipment: ", len_equipment)
    lenKT += len(range(len_muscles[0], len_muscles[1]))
    dfs = pd.DataFrame(columns=['musclesId', 'musclesName', 'equipmentId', 'equipmentName','image', 'gender', "ordering", "video_center", "video_side"])
    for i in range(len_muscles[0], len_muscles[1]):
        for j in range(len_equipment[0], len_equipment[1]):
            for k in range(len(ordering)):
                # print("i: ", i)
                # print("j: ", j)
                limit = 1
                url = f"https://musclewiki.com/newapi/exercise/exercises/?offset=0&limit={limit}&category={equipmen['id'][j]}&status={status}&ordering={ordering[k]}&muscles={muscles['id'][i]}"
                response = requests_api(url)
                if (response == False):
                    continue
                count = math.ceil(response['count'] / 50)
                for t in range(count):
                    limit = t * 50 + 50
                    if ((t + 1) == count):
                        limit = response['count']
                    url = f"https://musclewiki.com/newapi/exercise/exercises/?offset={t * 50}&limit={limit}&category={equipmen['id'][j]}&status={status}&ordering={ordering[k]}&muscles={muscles['id'][i]}"
                    # print(url)
                    data = requests_api(url)
                    data = data['results']
                    for l in range(limit - t * 50):
                        if (bool(data[l]['male_images']) != False):
                            video_male = {
                                'musclesId': muscles['id'][i],
                                'musclesName': muscles['name'][i],
                                'equipmentId': equipmen['id'][j],
                                'equipmentName': equipmen['name'][j],
                                'gender': 'male',
                                'image': data[l]['male_images'][0]['og_image'],
                                'ordering': ordering[k],
                                'video_center': None,
                                'video_side': None
                            }
                            if (len(data[l]['male_images']) == 2):
                                video_male['video_center'] = data[l]['male_images'][0]['branded_video']
                                video_male['video_side'] = data[l]['male_images'][1]['branded_video']
                            if (len(data[l]['male_images']) ==1):
                                if ('front.mp4' in data[l]['male_images'][0]['branded_video']):
                                    video_male['video_center'] = data[l]['male_images'][0]['branded_video']
                                else:
                                    video_male['video_side'] = data[l]['male_images'][0]['branded_video']
                            dfs.loc[len(dfs)] = video_male
                        video_female = None
                        if (bool(data[l]['female_images']) != False):
                            video_female = {
                                'musclesId': muscles['id'][i],
                                'musclesName': muscles['name'][i],
                                'equipmentId': equipmen['id'][j],
                                'equipmentName': equipmen['name'][j],
                                'gender': 'female',
                                'ordering': ordering[k],
                                'image': data[l]['female_images'][0]['og_image'],
                                'video_center': None,
                                'video_side': None
                            }
                        
                            if (len(data[l]['female_images']) == 2):
                                video_female['video_center'] = data[l]['female_images'][0]['branded_video']
                                video_female['video_side'] = data[l]['female_images'][1]['branded_video']
                            if (len(data[l]['female_images']) == 1):
                                if ('front.mp4' in data[l]['female_images'][0]['branded_video']):
                                    video_female['video_center'] = data[l]['female_images'][0]['branded_video']
                                else:
                                    video_female['video_side'] = data[l]['female_images'][0]['branded_video']
                            dfs.loc[len(dfs)] = video_female

        lenDF += len(dfs)
        print(lenDF)
    return dfs
```
